# Hello World

We will start  as almost all programming guides do, with the ubiquitous Hello World program. Hello World, for those of you who haven't seen it before, is a simple program that prints "Hello World" to the screen. It's a great place to get see some language basics as well as get a feel for what level the language is at. In general, the more that goes into writing a Hello World program, the lower the languae is and the closer the programmer is to the machine. With all that out of the way here is the program. Give it a good read over and we will break down what is going on below.

## The Program (hello_world.cpp)

```cpp
/*
Here is everyone's favorite teaching program, Hello World,
in the glorious langauge that is C++.
*/
#include <iostream> //iostream is the library for doing input and output

int main(){ // all programs begin at main
    std::cout << "Hello World!" << std::endl; // the actual printing
    return 0; //end main and return 0, which stands for program success
}
```

## Comments
Comments are ways for programmers to leave notes to themselves and other readers of their code. They are completly ignored by the compiler but are very imporant for writing quality code. In general comments should be used to 
- Explain why something was done
- Explain how to use the code
- Clarify what is being done, if what is being done is complicated

If what is being done, isn't complicated or is obvious from the code, don't bother commenting it. For example commenting `c = a + b` with "add a and b together and store the result in c" is a waste time to write and to read.

Comments in C++ come in two different flavors, single line and multiline.
### Single Line Comments
Single line comments begin with `//`. Everything after the `//` until the end of the line is treated as a comment and ignored by the compiler.
### Multiline Comments
Multiline comments are enclosed inside of `/* */`. Everything in between `/*` and `*/` will be ignored by the compiler. One of the interesting thing about multiline comments is that they can actually appear  inbetween pieces of code, though I don't recommend this as it is very hard to read.

Here is an example of making everyone's life harder

```cpp
#include <iostream>
std::cout << "Hello " /*an interruting comment*/ << "World!" /*that makes things harder to read*/ << std::endl;
```
To the compiler this just looks like `std::cout << "Hello "<< "World!" << std::endl;` but it is making your and my life harder.

## Preprocessor Directives
A preprocessor directive is any line that begins with a \#. It instructs the compiler to do something special on that line. We will cover #include here and the rest when you need them.

### #include \<filename>
#### What does #include do?
#include \<filename> means <u>copy and paste</u> the contents of **filename** into the current file. The big thing to pick up from here is that #include is just the C++ way of saying **copy and paste**. It does nothing more than that and thinking that it does more, can lead to you getting in trouble once programs get big. 

#### Why include iostream?
C++ was designed around the principle that you should only pay for what you use and so only the barest essentials are built into the langauge itself. Printing isn't one of those essentials but C++ comes with many standard libraries for taking care of common tasks. The library that deals with input and output is iostream so we must`#include <iostream>` at the top of our program if we want to be able to print.


[Here is a full list of all of libraries in the standard library](http://www.cplusplus.com/reference/) but don't dig too deep as you'll become quickly overwhelemed. 

#### What do the < > around the filename do?
They tell the compiler where to search for the file. `<>` means to search the standard system directories for the specified file. This means that all standard libaries, such as iostream, should be enclosed in `<>` as they live in the standard system directories on your computer. 


## What does int main(){ ... } do?
This line of code defines the main function of our program, with everything between the { } being what makes up the main function. Now there are some special things about the main function
1. Program execution always begins at main.
2. There must be exactly one function named main in the entire program.

So remember, regardless of what code appears before main, your program will always begin execution at main.

### What does the int in front of main mean?
It means that the main function will return an integer when the function is complete. You can see that main does so in its last line `return 0;`.

### Does whitespace matter?
For the most part, no, whitespace does not matter. The only rule for whitespace is that each preprocessor directive must be on its own line. This means that we could have written our program as
```cpp
#include <iostream>
int main(){std::cout<<"Hello World!"<<std::endl;return 0;}
```
and it will work just like the nicer looking version. 


Despite the fact that the compiler doesn't need the white space to work, I highly recommend that you use do use good whitespace and indentation in your code as it makes it a million times easier to read and understand.

### So if whitespace doesn't matter what denotes code belonging/ownership?
In C++, {} brackets denote belonging/ownership the same way that indentation does in Python. So `std::cout << "Hello World!" <<std::endl;` and `return 0;` belong to main and  make up main's body becuase they appear inside the {} that appear right after main's definition. 

### Can code appear outside of functions?
For the most part, no, code cannot appear outside of functions. The expections to this are preprocessor directives, and function, variable, class, struct, typedef, union, and enum declarations, and definitions. This means that all of your code logic (loops, if statements, function calls, etc.) will need to appear inside of a function definition.

## Can you break down the printing line?
Absolutely. The line again is `std::cout << "Hello World!" << std::endl;`
- std::cout: This is the variable that represents the standard output (the output to the terminal)
- <<: Is the out stream operator. We use this to write to std::cout and we use it to separate variables and expressions in a single write to the output.
- "Hello World!": This is the text we want to show on the screen. In C++ strings must be enclosed inside of double quotes. Single quotes do not work for strings (You can't do `'Hello World!'`. 
- std::endl: This is the end of the line and also causes the ostream (cout) that it is written to to be flushed 
  - To flush a stream means to empty the contents of the stream's memory buffer into the device it represents
  - In this case it means to make what we have written to cout to appear on the screen if it hasn't already
  
### What's up with the ; at the end of the line?
; mark the end of an expression in C++ and it must appear at the end of each and every expression in C++. Leaving them off is a syntax violation and your code won't be runnable if you forget them. 

#### What's an expression in C++?
Pretty much every line of code you write in C++ will be an expression. The only lines that don't need them are ones that end in either `{` or `}` but putting them there shouldn't hurt anything.


## What does return 0; do?
The return statement causes the function it appears inside to terminate and it gives back the value that comes after it. In this case, that value is 0. A return value of 0 from main means thatprogram completed successfully. A value other than 0 means that the program encountered some error with the exact error code being specific to each program.
